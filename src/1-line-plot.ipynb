```python
import matplotlib.pyplot as plt
import numpy as np

# Generating x values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculating y values as the square of x
y = x ** 2

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="y = x^2")
plt.title("Line Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
