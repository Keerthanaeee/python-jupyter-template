### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = x ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Creating the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Line chart
axs[0, 0].plot(x, y, label="y = x^2")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y")
axs[0, 0].legend()

# Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter)
for i, txt in enumerate(y_scatter):
    axs[1, 0].annotate(f'({x_scatter[i]}, {txt})', (x_scatter[i], y_scatter[i]))
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("y")

# Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
axs[1, 1].set_title("Pie Chart")
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
